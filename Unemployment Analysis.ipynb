```python
df2=df1[[' Estimated Unemployment Rate (%)',' Estimated Employed', ' Estimated Labour Participation Rate (%)' ]]
df2.head()
```

```python
#Plotting Heatmap
plt.figure(figsize=(5,5))
sns.heatmap(df2.corr(), annot=True, cmap='Reds', square=True, linewidths=1, linecolor='Black')
plt.show()
```

```python
# Analyzing Area wise Estimated Unemployment Rate
plt.figure(figsize=(5,4))
ax = sns.barplot(x='Area', y=' Estimated Unemployment Rate (%)', data=df1, palette='pastel')
ax.set_title("Area-Wise Estimated Unemployment Rate", fontsize=12)
plt.show()
```

```python
# Analyzing State wise Estimated Unemployment Rate
df1_sorted=df1.sort_values(by=" Estimated Unemployment Rate (%)")
plt.figure(figsize=(9,8))
ay=sns.barplot(x=" Estimated Unemployment Rate (%)", y="Region", data=df1_sorted)
ay.set_title("State-Wise Estimated Unemployment Rate", fontsize=12)
plt.show()
```

```python
#Analyzing Area-Wise Employment Rate
plt.figure(figsize=(8,6))
sns.histplot(x=" Estimated Employed", hue="Area", data=df1)
plt.show()
```

```python
#Analyzing State-Wise Employement Rate
plt.figure(figsize=(10,14))
sns.histplot(x=" Estimated Employed", hue="Region", data=df1, palette="Paired")
plt.show()
```

```python
# Analyzing highest rate of Unemployment by Year
group = df1.groupby([" Date"])[" Estimated Unemployment Rate (%)"].aggregate("max").reset_index()
plt.figure(figsize=(10,5))
az=sns.pointplot(x=group[' Date'].values, y=group[' Estimated Unemployment Rate (%)'].values)
plt.ylabel('Max rate (%)', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
